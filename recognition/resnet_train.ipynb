````python
# Configuration matching your project structure
DATA_ROOT = "LogoDet-3K"
ANNOTATION_FILE = os.path.join(DATA_ROOT, "annotations.json")

def setup_directories():
    """Create the required directory structure."""
    dirs = ['train', 'val', 'test']
    if os.path.exists(DATA_ROOT):
        print(f"Directory {DATA_ROOT} already exists. checking contents...")
    else:
        os.makedirs(DATA_ROOT)

    for d in dirs:
        os.makedirs(os.path.join(DATA_ROOT, d), exist_ok=True)
    print(f"‚úì Created directories in {DATA_ROOT}/")

def convert_bbox_to_coco(bbox):
    """
    Convert [xmin, ymin, xmax, ymax] to [x, y, w, h]
    """
    # Ensure bbox is a list/array
    if not isinstance(bbox, (list, tuple, np.ndarray)):
        return [0,0,0,0]

    x_min, y_min, x_max, y_max = bbox
    w = x_max - x_min
    h = y_max - y_min
    return [int(x_min), int(y_min), int(w), int(h)]

def download_and_process_dataset():
    setup_directories()
    try:
        print("‚è≥ Loading dataset from Hugging Face...")
        from datasets import load_dataset
        dataset = load_dataset("axonstan/LogoDet-3K", split="train")
    except Exception as e:
        print(f"‚ùå Error downloading dataset: {e}")
        return

    print("‚úì Dataset loaded. Starting processing and splitting...")

    # Initialize COCO structure
    coco_output = {
        "images": [],
        "categories": [],
        "annotations": []
    }

    # Create category mapping
    print("‚è≥ Building category map...")
    features = dataset.features
    if hasattr(features['company_name'], 'names'):
        category_names = features['company_name'].names
        category_map = {i: name for i, name in enumerate(category_names)}
    else:
        unique_ids = set(dataset['company_name'])
        category_map = {i: str(i) for i in unique_ids}

    for cat_id, cat_name in category_map.items():
        coco_output['categories'].append({
            "id": int(cat_id),
            "name": str(cat_name),
            "supercategory": "logo"
        })

    # SAMPLING: If IS_TEST is True, sample from the dataset
    total_images = len(dataset)

    if Config.IS_TEST:
        sample_size = min(Config.SAMPLE_SIZE, total_images)
        print(f"‚ö†Ô∏è  TEST MODE: Sampling {sample_size} from {total_images} images")
        indices = np.random.choice(total_images, sample_size, replace=False)
        indices.sort()  # Sort for consistent ordering
    else:
        indices = np.arange(total_images)
        print(f"Using full dataset: {total_images} images")

    # NOW split train/val/test from the SAMPLED indices
    np.random.seed(42)
    np.random.shuffle(indices)

    n_samples = len(indices)
    train_split = int(0.8 * n_samples)
    val_split = int(0.9 * n_samples)

    print(f"Sample size: {n_samples}")
    print(f"Splits: Train: {train_split}, Val: {val_split-train_split}, Test: {n_samples-val_split}")

    # Process images
    ann_id_counter = 1

    for array_idx, dataset_idx in enumerate(tqdm(indices, desc="Saving images & annotations")):
        item = dataset[int(dataset_idx)]

        # Determine split based on position in SAMPLED array
        if array_idx < train_split:
            split = 'train'
        elif array_idx < val_split:
            split = 'val'
        else:
            split = 'test'

        # Image handling
        image = item['image_path']
        file_name = f"{split}_{dataset_idx:06d}.jpg"
        save_path = os.path.join(DATA_ROOT, split, file_name)

        # Save image if it doesn't exist
        if not os.path.exists(save_path):
            try:
                if image.mode != 'RGB':
                    image = image.convert('RGB')
                image.save(save_path)
            except Exception as e:
                print(f"Error saving {file_name}: {e}")
                continue

        w, h = image.size

        # Add Image info to COCO
        coco_output["images"].append({
            "id": int(dataset_idx),
            "file_name": file_name,
            "width": w,
            "height": h,
            "split": split
        })

        # Add Annotation info
        bbox_orig = item['bbox']
        bbox_coco = convert_bbox_to_coco(bbox_orig)
        area = bbox_coco[2] * bbox_coco[3]

        coco_output["annotations"].append({
            "id": ann_id_counter,
            "image_id": int(dataset_idx),
            "category_id": int(item['company_name']),
            "bbox": bbox_coco,
            "area": area,
            "iscrowd": 0
        })
        ann_id_counter += 1

    # Save Annotations File
    print(f"‚è≥ Saving annotations to {ANNOTATION_FILE}...")
    with open(ANNOTATION_FILE, 'w') as f:
        json.dump(coco_output, f)

    print("\n Dataset preparation complete!")
    print(f"Images saved to: {DATA_ROOT}/[train|val|test]")
    print(f"Annotations saved to: {ANNOTATION_FILE}")

if __name__ == "__main__":
    # Check if we need to run
    if os.path.exists(ANNOTATION_FILE):
        print("Dataset appears to be present already.")
        choice = input("Re-download and overwrite? (y/n): ")
        if choice.lower() == 'y':
            download_and_process_dataset()
    else:
        download_and_process_dataset()
````

````python
def get_transforms(split='train', img_size=224):
    """
    Get augmentation transforms for training/validation/testing.
    Uses Albumentations for more advanced augmentations.
    """
    if split == 'train':
        return A.Compose([
            A.Resize(img_size, img_size),
            A.HorizontalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.15, rotate_limit=15, p=0.5),
            A.OneOf([
                A.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1, p=1.0),
                A.RandomBrightnessContrast(brightness_limit=0.2, contrast_limit=0.2, p=1.0),
            ], p=0.5),
            A.GaussNoise(var_limit=(10.0, 50.0), p=0.2),
            A.GaussianBlur(blur_limit=(3, 5), p=0.2),
            A.Normalize(
                mean=[0.485, 0.456, 0.406],
                std=[0.229, 0.224, 0.225]
            ),
            ToTensorV2()
        ])
    else:
        return A.Compose([
            A.Resize(img_size, img_size),
            A.Normalize(
                mean=[0.485, 0.456, 0.406],
                std=[0.229, 0.224, 0.225]
            ),
            ToTensorV2()
        ])

print("‚úì Augmentation pipelines defined")
````

````python
class LogoDataset(Dataset):
    """
    Custom Dataset for LogoDet-3K logo recognition.
    Loads cropped logo regions based on bounding box annotations.
    """
    def __init__(self, df, data_root, split='train', transform=None, max_samples_per_class=None):
        self.df = df.copy()
        self.data_root = data_root
        self.split = split
        self.transform = transform

        # Filter by split
        self.df = self.df[self.df['split'] == split].reset_index(drop=True)

        # Apply class balancing if specified
        if max_samples_per_class is not None:
            self.df = self._balance_classes(max_samples_per_class)

        # Create label encoding
        self.brand_to_idx = {brand: idx for idx, brand in enumerate(sorted(self.df['brand_name'].unique()))}
        self.idx_to_brand = {idx: brand for brand, idx in self.brand_to_idx.items()}
        self.num_classes = len(self.brand_to_idx)

        # Add numeric labels
        self.df['label'] = self.df['brand_name'].map(self.brand_to_idx)

        print(f"\n{split.upper()} Dataset:")
        print(f"  Samples: {len(self.df):,}")
        print(f"  Classes: {self.num_classes}")

    def _balance_classes(self, max_samples):
        """Cap the number of samples per class to handle class imbalance."""
        balanced_dfs = []
        for brand in self.df['brand_name'].unique():
            brand_df = self.df[self.df['brand_name'] == brand]
            if len(brand_df) > max_samples:
                brand_df = brand_df.sample(n=max_samples, random_state=42)
            balanced_dfs.append(brand_df)
        return pd.concat(balanced_dfs, ignore_index=True)

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]

        # Construct image path based on split
        img_path = os.path.join(self.data_root, self.split, row['image_filename'])

        # Load image
        try:
            image = cv2.imread(img_path)
            image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        except Exception as e:
            print(f"Error loading {img_path}: {e}")
            # Return a blank image if loading fails
            image = np.zeros((224, 224, 3), dtype=np.uint8)

        # Crop logo region using bbox [x, y, w, h]
        bbox = row['bbox']
        x, y, w, h = [int(v) for v in bbox]

        # Ensure bbox is within image bounds
        img_h, img_w = image.shape[:2]
        x = max(0, min(x, img_w - 1))
        y = max(0, min(y, img_h - 1))
        w = max(1, min(w, img_w - x))
        h = max(1, min(h, img_h - y))

        # Crop
        logo = image[y:y+h, x:x+w]

        # Handle empty crops
        if logo.size == 0:
            logo = np.zeros((224, 224, 3), dtype=np.uint8)

        # Apply transforms
        if self.transform:
            logo = self.transform(image=logo)['image']

        label = row['label']

        return logo, label

    def get_class_weights(self):
        """Compute class weights for handling imbalance."""
        class_counts = self.df['label'].value_counts().sort_index()
        weights = 1.0 / class_counts.values
        weights = weights / weights.sum() * len(weights)  # Normalize
        return torch.FloatTensor(weights)

print("‚úì Dataset class defined")
````

````python
class ArcFaceLoss(nn.Module):
    """
    ArcFace Loss: Additive Angular Margin Loss for Deep Face Recognition
    Reference: https://arxiv.org/abs/1801.07698
    """
    def __init__(self, embedding_dim, num_classes, s=64.0, m=0.5):
        super(ArcFaceLoss, self).__init__()
        self.embedding_dim = embedding_dim
        self.num_classes = num_classes
        self.s = s  # Scale parameter
        self.m = m  # Angular margin

        # Weight matrix for classification
        self.weight = nn.Parameter(torch.FloatTensor(num_classes, embedding_dim))
        nn.init.xavier_uniform_(self.weight)

        self.cos_m = math.cos(m)
        self.sin_m = math.sin(m)
        self.threshold = math.cos(math.pi - m)
        self.mm = math.sin(math.pi - m) * m

    def forward(self, embeddings, labels):
        """
        Args:
            embeddings: (batch_size, embedding_dim) - L2 normalized embeddings
            labels: (batch_size,) - ground truth labels
        Returns:
            loss: scalar
        """
        # Normalize weight matrix
        weight_norm = F.normalize(self.weight, p=2, dim=1)

        # Cosine similarity
        cosine = F.linear(F.normalize(embeddings), weight_norm)

        # Get cosine for ground truth class
        sine = torch.sqrt(1.0 - torch.pow(cosine, 2))
        phi = cosine * self.cos_m - sine * self.sin_m

        # Apply threshold
        phi = torch.where(cosine > self.threshold, phi, cosine - self.mm)

        # One-hot encoding
        one_hot = torch.zeros(cosine.size(), device=embeddings.device)
        one_hot.scatter_(1, labels.view(-1, 1).long(), 1)

        # Apply margin to ground truth class
        output = (one_hot * phi) + ((1.0 - one_hot) * cosine)
        output *= self.s

        # Cross entropy loss
        loss = F.cross_entropy(output, labels)

        return loss, output

print("‚úì ArcFace loss defined")
````

````python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torchvision import models


class LogoRecognitionModel(nn.Module):
    """
    ResNet50 backbone + linear classifier cho LogoDet3K.
    Tr·∫£ v·ªÅ:
      - loss  (n·∫øu labels kh√¥ng None)
      - logits
      - embeddings (feature vector ƒë·ªÉ d√πng sau n√†y n·∫øu mu·ªën ArcFace / retrieval)
    """
    def __init__(self, num_classes: int, pretrained: bool = True):
        super().__init__()

        # Backbone ResNet50
        if pretrained:
            backbone = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V2)
        else:
            backbone = models.resnet50(weights=None)

        in_features = backbone.fc.in_features
        backbone.fc = nn.Identity()  # b·ªè fc g·ªëc

        self.backbone = backbone
        self.embedding = nn.Linear(in_features, 512)
        self.bn = nn.BatchNorm1d(512)
        self.dropout = nn.Dropout(0.3)
        self.classifier = nn.Linear(512, num_classes)

        # Loss d√πng CrossEntropy v·ªõi label smoothing cho ‚Äúm∆∞·ª£t‚Äù
        self.criterion = nn.CrossEntropyLoss(label_smoothing=0.1)

    def forward(self, x, labels=None):
        # Backbone feature
        feat = self.backbone(x)           # [B, 2048]
        emb = self.embedding(feat)        # [B, 512]
        emb = self.bn(emb)
        emb = F.relu(emb)
        emb = self.dropout(emb)

        logits = self.classifier(emb)     # [B, num_classes]

        if labels is not None:
            loss = self.criterion(logits, labels)
            return loss, logits, emb
        else:
            return logits, emb
````

````python
def train_epoch(model, loader, optimizer, scheduler, scaler, epoch, device):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    pbar = tqdm(loader, desc=f"Epoch {epoch+1}/{Config.NUM_EPOCHS} [Train]")
    for images, labels in pbar:
        images = images.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()

        if Config.USE_AMP:
            with torch.cuda.amp.autocast():
                loss, logits, _ = model(images, labels)
        else:
            loss, logits, _ = model(images, labels)

        if Config.USE_AMP:
            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()
        else:
            loss.backward()
            optimizer.step()

        if scheduler is not None:
            scheduler.step()

        running_loss += loss.item() * images.size(0)

        # ‚úÖ accuracy d·ª±a tr√™n logits softmax b√¨nh th∆∞·ªùng
        _, preds = torch.max(logits, 1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)

        pbar.set_postfix(
            loss=running_loss / total,
            acc=100.0 * correct / total
        )

    epoch_loss = running_loss / total
    epoch_acc = 100.0 * correct / total
    return epoch_loss, epoch_acc
````

````python
from sklearn.metrics import f1_score

def validate(model, loader, device):
    model.eval()
    running_loss = 0.0
    correct = 0
    total = 0
    all_preds = []
    all_labels = []

    with torch.no_grad():
        pbar = tqdm(loader, desc="Validation")
        for images, labels in pbar:
            images = images.to(device)
            labels = labels.to(device)

            # forward
            loss, logits, _ = model(images, labels)

            running_loss += loss.item() * images.size(0)

            _, preds = torch.max(logits, 1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)

            all_preds.extend(preds.cpu().tolist())
            all_labels.extend(labels.cpu().tolist())

            pbar.set_postfix(
                loss=running_loss / total,
                acc=100.0 * correct / total
            )

    epoch_loss = running_loss / total
    epoch_acc = 100.0 * correct / total
    f1 = f1_score(all_labels, all_preds, average="macro", zero_division=0)
    return epoch_loss, epoch_acc, f1
````

````python
def save_checkpoint(model, optimizer, scheduler, epoch, val_acc, is_best, force_save=False):
    """
    Save checkpoint with configurable frequency.
    """
    save_interval = 10
    
    should_save = (
        force_save or
        is_best or
        (epoch + 1) % save_interval == 0 or
        (epoch + 1) == Config.NUM_EPOCHS
    )
    
    if not should_save:
        if is_best:
            print(f"  (‚≠ê New Best: {val_acc:.2f}% - will save at epoch {((epoch+1)//10+1)*10})")
        return
    
    print(f"  üíæ Saving checkpoint (epoch {epoch+1})...", end=" ", flush=True)

    # Convert Config to dict
    config_dict = {k: v for k, v in Config.__dict__.items() 
                   if not k.startswith('__') and not callable(v)}

    # Create checkpoint with history
    checkpoint = {
        'epoch': epoch,
        'state_dict': model.state_dict(),
        'optimizer': optimizer.state_dict(),
        'scheduler': scheduler.state_dict() if scheduler else None,
        'val_acc': val_acc,
        'config': config_dict,
        'history': history  # ‚úÖ ADDED: Save history in checkpoint
    }

    # Save last checkpoint
    last_path = os.path.join(Config.CHECKPOINT_DIR, 'last_checkpoint.pth')
    torch.save(checkpoint, last_path)

    # Save best checkpoint
    if is_best:
        best_path = os.path.join(Config.CHECKPOINT_DIR, 'best_checkpoint.pth')
        torch.save(checkpoint, best_path)
        print(f"Done. (‚≠ê New Best: {val_acc:.2f}%)")
    else:
        if (epoch + 1) % save_interval == 0:
            epoch_path = os.path.join(Config.CHECKPOINT_DIR, f'checkpoint_epoch_{epoch+1}.pth')
            torch.save(checkpoint, epoch_path)
        print("Done.")
````

````python
from sklearn.metrics import accuracy_score, top_k_accuracy_score, f1_score
import numpy as np

def evaluate_comprehensive(model, loader, device, dataset_name='Test'):
    """
    Comprehensive evaluation with multiple metrics.
    """
    model.eval()

    all_preds = []
    all_labels = []
    all_probs = []

    print(f"\nEvaluating on {dataset_name} set...")

    with torch.no_grad():
        for images, labels in tqdm(loader, desc=f"{dataset_name} Evaluation"):
            images = images.to(device)
            labels = labels.to(device)

            if Config.USE_AMP:
                with torch.cuda.amp.autocast():
                    _, logits, _ = model(images, labels)
            else:
                _, logits, _ = model(images, labels)

            probs = F.softmax(logits, dim=1)
            _, predicted = logits.max(1)

            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
            all_probs.extend(probs.cpu().numpy())

    all_preds = np.array(all_preds)
    all_labels = np.array(all_labels)
    all_probs = np.array(all_probs)

    # ---------------------------------------------------------
    # üîß FIX: Define the full list of classes (0 to 2999)
    # This prevents the error when testing on small subsets
    # ---------------------------------------------------------
    all_classes = np.arange(Config.NUM_CLASSES)

    # Calculate metrics
    top1_acc = accuracy_score(all_labels, all_preds) * 100

    # Top-k accuracy (Added 'labels=all_classes')
    top5_acc = top_k_accuracy_score(all_labels, all_probs, k=5, labels=all_classes) * 100
    top10_acc = top_k_accuracy_score(all_labels, all_probs, k=10, labels=all_classes) * 100

    # F1 scores
    f1_macro = f1_score(all_labels, all_preds, average='macro', zero_division=0)
    f1_weighted = f1_score(all_labels, all_preds, average='weighted', zero_division=0)

    # Print results
    print("\n" + "="*70)
    print(f"{dataset_name.upper()} SET RESULTS")
    print("="*70)
    print(f"\n Accuracy Metrics:")
    print(f"  Top-1 Accuracy:  {top1_acc:.2f}%")
    print(f"  Top-5 Accuracy:  {top5_acc:.2f}%")
    print(f"  Top-10 Accuracy: {top10_acc:.2f}%")
    print(f"\n F1 Scores:")
    print(f"  F1 (Macro):    {f1_macro:.4f}")
    print(f"  F1 (Weighted): {f1_weighted:.4f}")

    results = {
        'top1_acc': top1_acc,
        'top5_acc': top5_acc,
        'top10_acc': top10_acc,
        'f1_macro': f1_macro,
        'f1_weighted': f1_weighted,
        'predictions': all_preds,
        'labels': all_labels,
        'probabilities': all_probs
    }

    return results

# Evaluate on test set
test_results = evaluate_comprehensive(model, test_loader, device, 'Test')
````

````python
# =========================================================
# üì¶ CREATE & LOAD MODEL - CORRECT ARCHITECTURE
# =========================================================

import torch
import torch.nn as nn
import torch.nn.functional as F
from torchvision import models
import os

print("="*70)
print("üî® CREATING MODEL WITH CORRECT ARCHITECTURE")
print("="*70)

# ƒê·ªãnh nghƒ©a model gi·ªëng nh∆∞ trong notebook
class LogoRecognitionModel(nn.Module):
    """
    ResNet50 backbone + linear classifier cho LogoDet3K.
    """
    def __init__(self, num_classes: int, pretrained: bool = True):
        super().__init__()

        # Backbone ResNet50
        if pretrained:
            backbone = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V2)
        else:
            backbone = models.resnet50(weights=None)

        in_features = backbone.fc.in_features
        backbone.fc = nn.Identity()  # b·ªè fc g·ªëc

        self.backbone = backbone
        self.embedding = nn.Linear(in_features, 512)
        self.bn = nn.BatchNorm1d(512)
        self.dropout = nn.Dropout(0.3)
        self.classifier = nn.Linear(512, num_classes)

        # Loss d√πng CrossEntropy v·ªõi label smoothing
        self.criterion = nn.CrossEntropyLoss(label_smoothing=0.1)

    def forward(self, x, labels=None):
        # Backbone feature
        feat = self.backbone(x)           # [B, 2048]
        emb = self.embedding(feat)        # [B, 512]
        emb = self.bn(emb)
        emb = F.relu(emb)
        emb = self.dropout(emb)

        logits = self.classifier(emb)     # [B, num_classes]

        if labels is not None:
            loss = self.criterion(logits, labels)
            return loss, logits, emb
        else:
            return logits, emb

# L·∫•y num_classes
if hasattr(Config, 'NUM_CLASSES') and Config.NUM_CLASSES is not None:
    num_classes = Config.NUM_CLASSES
else:
    num_classes = train_dataset.num_classes if 'train_dataset' in globals() else 3000
    Config.NUM_CLASSES = num_classes

print(f"‚úÖ Creating model with {num_classes} classes")

# Kh·ªüi t·∫°o model
model = LogoRecognitionModel(num_classes=num_classes, pretrained=False)

print("="*70)
print("üîç LOADING CHECKPOINT")
print("="*70)

CHECKPOINT_PATH = "outputs/checkpoints/best_checkpoint.pth"

if os.path.exists(CHECKPOINT_PATH):
    file_size = os.path.getsize(CHECKPOINT_PATH) / (1024*1024)
    print(f"‚úÖ Found checkpoint: {CHECKPOINT_PATH}")
    print(f"üìä File size: {file_size:.2f} MB")
    
    try:
        # Load checkpoint
        checkpoint = torch.load(CHECKPOINT_PATH, map_location=device, weights_only=False)
        print("‚úÖ Checkpoint loaded successfully!")
        
        # Load state_dict
        if 'state_dict' in checkpoint:
            model.load_state_dict(checkpoint['state_dict'])
            print(f"\n‚úÖ Loaded model weights from epoch {checkpoint.get('epoch', 'unknown') + 1}")
            
            # Hi·ªÉn th·ªã th√¥ng tin
            if 'val_acc' in checkpoint:
                print(f"üìä Validation accuracy: {checkpoint['val_acc']:.2f}%")
            if 'config' in checkpoint:
                print(f"üìã NUM_CLASSES in checkpoint: {checkpoint['config'].get('NUM_CLASSES', 'N/A')}")
        else:
            print("‚ö†Ô∏è 'state_dict' key not found in checkpoint")
        
        # Move to device v√† eval mode
        model = model.to(device)
        model.eval()
        print(f"\n‚úÖ Model ready on {device}")
        
        # Count parameters
        total_params = sum(p.numel() for p in model.parameters())
        print(f"üìä Total parameters: {total_params:,}")
        print("="*70)
        
    except Exception as e:
        print(f"\n‚ùå Error loading checkpoint: {str(e)}")
        import traceback
        traceback.print_exc()
else:
    print(f"‚ùå Checkpoint not found: {CHECKPOINT_PATH}")
````

````python
# =========================================================
# üß™ TEST MANUALLY
# =========================================================

import torchvision.transforms as transforms
from PIL import Image
import matplotlib.pyplot as plt
import os

def test_single_image(image_path, model, train_dataset, device, top_k=5):
    """Test model manually with a single image"""
    
    transform = transforms.Compose([
        transforms.Resize((Config.IMG_SIZE, Config.IMG_SIZE)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], 
                           std=[0.229, 0.224, 0.225])
    ])
    
    try:
        # Load ·∫£nh
        image = Image.open(image_path).convert('RGB')
        original_image = image.copy()
        
        # Transform
        input_tensor = transform(image).unsqueeze(0).to(device)
        
        # Inference
        model.eval()
        with torch.no_grad():
            logits, emb = model(input_tensor)
            probs = torch.softmax(logits, dim=1)
            top_probs, top_indices = torch.topk(probs, top_k, dim=1)
            
            top_probs = top_probs[0].cpu().numpy()
            top_indices = top_indices[0].cpu().numpy()
        
        # Visualize
        fig, axes = plt.subplots(1, 2, figsize=(15, 5))
        
        axes[0].imshow(original_image)
        axes[0].axis('off')
        axes[0].set_title(f'Input: {os.path.basename(image_path)}', fontsize=12, fontweight='bold')
        
        axes[1].barh(range(top_k), top_probs, color='steelblue')
        axes[1].set_yticks(range(top_k))
        
        brand_names = []
        for idx in top_indices:
            if idx in train_dataset.idx_to_brand:
                brand_names.append(train_dataset.idx_to_brand[idx])
            else:
                brand_names.append(f"Unknown_{idx}")
        
        axes[1].set_yticklabels([f"{name}\n({prob:.1%})" 
                                 for name, prob in zip(brand_names, top_probs)])
        axes[1].invert_yaxis()
        axes[1].set_xlabel('Confidence', fontsize=11)
        axes[1].set_title(f'Top {top_k} Predictions', fontsize=12, fontweight='bold')
        axes[1].grid(axis='x', alpha=0.3, linestyle='--')
        axes[1].set_xlim(0, 1)
        
        plt.tight_layout()
        plt.show()
        
        print("\n" + "="*70)
        print(" PREDICTION RESULTS")
        print("="*70)
        for i, (brand, prob) in enumerate(zip(brand_names, top_probs), 1):
            emoji = "ü•á" if i == 1 else "ü•à" if i == 2 else "ü•â" if i == 3 else "  "
            print(f"{emoji} {i}. {brand:35s} {prob:>6.2%}")
        print("="*70)
        
        return brand_names[0], top_probs[0]
        
    except Exception as e:
        print(f"‚ùå Error: {str(e)}")
        import traceback
        traceback.print_exc()

# =========================================================
#  RUN TEST
# =========================================================

IMAGE_PATH = "LogoDet-3K/train/train_067188.jpg"  # Change path here

if 'model' in globals() and 'train_dataset' in globals():
    print(f" Testing: {IMAGE_PATH}\n")
    if os.path.exists(IMAGE_PATH):
        test_single_image(IMAGE_PATH, model, train_dataset, device, top_k=5)
    else:
        print(f"‚ùå File does not exist: {IMAGE_PATH}")
else:
    print(" Please run the cell to load the model first!")
````
